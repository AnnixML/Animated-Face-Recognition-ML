```python
import pandas as pd
import tkinter as tk
from PIL import ImageTk, Image
import os


df = pd.read_csv('image_tags.csv')

def display_image_and_label(index):
    row = df.iloc[index]
    image_path = row[0]
    label = row[1]

    img = Image.open(image_path)
    img = img.resize((400, 400))  

    root = tk.Tk()
    root.title("Image and Label")

    photo = ImageTk.PhotoImage(img)
    image_label = tk.Label(root, image=photo)
    image_label.photo = photo 
    image_label.pack()

    text_label = tk.Label(root, text=f"Label: {label}", font=("Arial", 16))
    text_label.pack(pady=10)
    text_label = tk.Label(root, text=f"Path: {row[0]}", font=("Arial", 16))
    text_label.pack(pady=10)

    button_frame = tk.Frame(root)
    button_frame.pack(pady=10)

    keep_button = tk.Button(button_frame, text="Keep", command=lambda: keep_row(index, root))
    keep_button.pack(side=tk.LEFT, padx=10)

    delete_button = tk.Button(button_frame, text="Delete", command=lambda: delete_row(index, root, image_path))
    delete_button.pack(side=tk.LEFT, padx=10)

    root.mainloop()

def delete_row(index, root, path):
    root.destroy()

    global df
    df = df.drop(index).reset_index(inplace=True)
    df.to_csv('image_tags.csv', index=False)
    try:
        os.remove(path)
        print(f"Deleted image file: {path}")
    except OSError as e:
        print(f"Error deleting image file: {e.strerror}")
    if current_index < len(df):
        display_image_and_label(current_index)
    else:
        print("No more rows to display.")
def keep_row(index, root):
    root.destroy()

    with open('index.txt', 'w') as f:
        f.write(str(index + 1))

    global current_index
    current_index = index + 1

    if current_index < len(df):
        display_image_and_label(current_index)
    else:
        print("No more rows to display.")
```

```python
f = open("index.txt", "r")
current_index=int(f.read())
display_image_and_label(current_index)
```
